# what is open CV

OpenCV (Open Source Computer Vision Library) is a popular library used for image processing, computer vision, and AI camera applications.

It helps computers:

Read images
Detect faces
Recognize objects
Process videos
Track motion
Work with webcams and CCTV feeds

It is mainly used with Python and C++.

What Can OpenCV Do?
Image Processing
Resize images
Crop images
Blur/sharpen
Convert color (RGB → Gray)
Face Detection

Example:

Phone face unlock
Attendance systems
Video Processing
Read webcam
Detect movement
Count objects
AI & Machine Learning Projects
Self-driving cars
Gesture control
OCR (text reading from images)
Drawing & Graphics
Draw shapes
Add text on images

# Core OpenCV Modules

| Module           | Purpose                                  |
| ---------------- | ---------------------------------------- |
| `cv2.core`       | Basic operations, matrices, arrays       |
| `cv2.imgproc`    | Image processing functions               |
| `cv2.imgcodecs`  | Read/write image files                   |
| `cv2.highgui`    | Display images & GUI windows             |
| `cv2.videoio`    | Camera and video handling                |
| `cv2.video`      | Motion tracking & background subtraction |
| `cv2.calib3d`    | Camera calibration & 3D vision           |
| `cv2.features2d` | Feature detection algorithms             |
| `cv2.objdetect`  | Face/object detection                    |
| `cv2.dnn`        | Deep Neural Network module               |
| `cv2.ml`         | Machine learning algorithms              |
| `cv2.photo`      | Computational photography                |
| `cv2.stitching`  | Panorama image stitching                 |
| `cv2.flann`      | Fast nearest-neighbor search             |
| `cv2.gapi`       | Graph API for optimized pipelines        |


# Read and show image

```python
import cv2 as cv
import numpy as np
import os   
```

```python
# read image
img=cv.imread("E:\Open_Cv\img\img_2.jpg")
cv.imshow("rama",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

# resize image

```python
# read image
img=cv.imread("E:\Open_Cv\img\img_2.jpg")
re_img=cv.resize(img,(300,200))
cv.imshow("rama",re_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

# multiple image show in one frame

```python
# read image
img=cv.imread("E:\Open_Cv\img\img_2.jpg")
re_img=cv.resize(img,(300,200))
h=np.hstack((re_img,re_img))
cv.imshow("rama",h)
cv.waitKey(0)
cv.destroyAllWindows()
```
